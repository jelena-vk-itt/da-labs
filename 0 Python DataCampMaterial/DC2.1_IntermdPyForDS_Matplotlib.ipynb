### Scatterplot

At this stage typing out arrays of data for use in our examples is becoming tedious. We will, instead, get some data that is available on the web. See notebook [A_GettingDataFromAnExternalFile](A_GettingDataFromAnExternalFile.html). We will read historical pole position (latitude, longitude) data as explained in that notebook. 

```python
from numpy import genfromtxt
polar_data = genfromtxt('polar.csv', delimiter=',')   # R sets: polar
polar_data = polar_data[1:, 1:]
print(polar_data)
```

Now we plot the longitudes aganst the latitudes:

```python
plt.plot(polar_data[:,0], polar_data[:,1])
plt.show()
```

The plot doesn't look great. That is because the values are not ordered and, what is more, ordering them by value would not make much sense (why?). A more suitable plot in this case is the scatterplot.

```python
plt.scatter(polar_data[:,0], polar_data[:,1])
plt.show()
```

### Histogram

Bins can be specified. 

Here we introduce the function `plt.clf()`, which allows us to clear the plot that was shown and start afresh in building the next one.

```python
plt.hist(polar_data[:,0], bins=5)
plt.show()
plt.clf()
plt.hist(polar_data[:,1])
plt.show()
plt.clf()
```

### Boxplot

```python
plt.boxplot(polar_data)
```

## Plot customisation

Customisation is not a trivial task with respect to data analysis. Conveying a data analytical message properly often depends on these details, which have to be chosen carefully.

```python
plt.scatter(polar_data[:,0], polar_data[:,1])
# plt.yscale('log') 

# Strings
xlab = 'Polar latitude'
ylab = 'Polar longigude'
title = 'Pole Positions of New Caledonian Laterites (R Dataset)'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)
# Add title
plt.title(title)

tick_val = [-100, -50, 0]
tick_lab = ['-100', '-50', '0']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)

# After customizing, display the plot
plt.show()
```

Now we are going to add information about the sequece of polar positions, representing them with increasingly larger dots in order of time. Let's have a look at how to create a sequence of numbers first.

```python
rng = range(3)
for i in rng:
    print(i)
print("-------------")
rng2 = range(1, 4, 1)
for i in rng2:
    print(i)
print("-------------")
rng3 = range(1, 11, 2)
for i in rng3:
    print(i)
```

Now we define a list of sizes for the 50 data points and use them with the `s` attribute to make the dots increase in size as the plot progresses through the data points (ordered by time).

```python
dot_sizes = range(1, 51)
plt.scatter(polar_data[:,0], polar_data[:,1], s=dot_sizes)
plt.show()
```

For a clearer representation, we can make the dots grow a bit more:

```python
dot_sizes = range(1, 100, 2)
plt.scatter(polar_data[:,0], polar_data[:,1], s=dot_sizes)
plt.show()
```

We can colour the dots in the same way by passing an argument for parameter `c`. The same piece of code below demostrates how some text can be added anywhere in the plot.

```python
colours = ['red'] * 10 + ['yellow'] * 10 + ['green'] * 10 + ['blue'] * 10 + ['black'] * 10 # define the colour list
colours
```

```python
dot_sizes = range(1, 100, 2)
plt.scatter(polar_data[:,0], polar_data[:,1], s=dot_sizes, c=colours)
plt.text(-60, 200, 'Colours everywhere!')
plt.show()
```
