```python
import numpy as np
def categoricalNB (feature_df, target_arr):
    cnb = {"Priors":pd.DataFrame({'Probability':[]})}
    overall_count = len(target_arr)
    for tgtcls in set(target_arr):
        outcome_count = (target_arr == tgtcls).value_counts().reindex([True, False], fill_value = 0)[True]
        probability_estimate = outcome_count/overall_count
        cnb['Priors'].loc[tgtcls] = [ probability_estimate]
    
    for feature in feature_df.columns:
        target_value_set = list(set(target_arr))
        cnb[feature] = pd.DataFrame(columns = target_value_set)
        for value in set(feature_df[feature]):
            probability_estimates = []
            for tgtcls in target_value_set:
                evidence_given_outcome_count = np.logical_and(feature_df[feature] == value, target_arr == tgtcls). \
                                                  value_counts().reindex([True, False], fill_value = 0)[True]
                outcome_count = (target_arr == tgtcls).value_counts().reindex([True, False], fill_value = 0)[True]
                probability_estimates.append(evidence_given_outcome_count/outcome_count)
            cnb[feature].loc[value] = probability_estimates    
            
    return cnb

categoricalNB(play_df.iloc[:,:-1], play_df['Play'])
```
