# Reading data into a NumPy array

```python
from numpy import genfromtxt
polar_data = genfromtxt('polar.csv', delimiter=',')   # R sets: polar
polar_data
```

As we are working with NumPy arrays only, we are limited to a single data type, which in this case is `float`. We will 'get rid of' the empty array positions, which resulted from non-numeric values in the CSV file.

```python
polar_data = polar_data[1:, 1:]
polar_data
```
